```python
# Part 3 - Describe what an identity matrix is. What does multiplying with an identity matrix do?

# Identity Matrix has following property:
# - It is a square matrix, meaning it has the same number of rows and columns.
# - All of its diagonal elements are equal to 1, while all other elements are equal to 0.
# - When we multiply any matrix by an identity matrix of the appropriate size, the result is the original matrix itself.
# - The identity matrix acts as a neutral element for matrix multiplication.
```

```python
# Part 3 - Create an identity matrix called identity_matrix that has the same size as img_average_channel.

# Setting a varaible img_average_channel_shape to get the shape of img_average_channel
img_average_channel_shape = img_average_channel.shape

# Create an identity matrix of the same size
identity_matrix = np.identity(img_average_channel_shape[0])

# Print the identity matrix
print(identity_matrix)
```

```python
# Part 3 - Print a random row out of identity_matrix and describe your observations.

# Lets get the shape of the identity_matrix
identity_matrix_shape = identity_matrix.shape

# Generate a random row index within the valid range
random_row_index = np.random.randint(0, identity_matrix_shape[0])

# Extract and print the random row from the identity_matrix
random_row = identity_matrix[random_row_index]
print("Random Row from Identity Matrix:" , random_row)


# The position of the 1 in the random row will change each time we run the code due to the randomness of the row selection and
#all other elements in the row will be 0, as per the definition of an identity matrix.
```

```python
# Part -4 First, create a numpy array that contains the numbers 0,1,2,......n-1 where n  is the number of rows in the image.

# Lets get the number of rows in the image (should be 100)
n = image.shape[0]
print(n)

# Create a numpy array containing numbers from 0 to n-1 using np
numbers_array = np.arange(n)

# Print the numbers_array
print(numbers_array)
```

```python
# Part 4 - Use numpy.random.shuffle to shuffle the indices. Call the shuffled array, rows

# Shuffle the indices
np.random.shuffle(numbers_array)

# Store the shuffled array as "rows"
rows = numbers_array
```

```python
# Part 4 - Print rows and inspect the elements.

print(rows) #(we notice that the numbers have been shuffled)
```

```python
# Part 5 - Use your array of random row indices to create a new matrix T from identity_matrix. 
#Overall, T has the same elements as identity_matrix, but the order of the rows of T should be according to the variable rows

# Create a new matrix T by rearranging rows according to the "rows" variable
T = identity_matrix[rows]

# Print the matrix T
print("Matrix T:")
print(T)

#Each row in T represents one of the rows from the identity_matrix, but the order in which these rows appear in T is randomized.
```

```python
# part 5 - Sanity check, print the shape of T and make sure it is the same size as identity_matrix and img_average_channel. If you need to, reshape it to have the same size.

print(identity_matrix.shape)
print(img_average_channel.shape)
print(T.shape)
```

```python
# Part - 6 - Matrix multiply T and img_average_channel and store the result in transformed_image. Check the shape of the result.

transformed_image = np.dot(T, img_average_channel)

# To Check the shape of the result
print("Shape of transformed_image:", transformed_image.shape)

#The shape of transformed_image should depend on the sizes of T and img_average_channel, and it will be consistent with the rules of matrix multiplication
```

```python
# Part 6 - Plot transformed_image and describe the pattern you see.

plt.imshow(transformed_image) 
plt.title('Transformed Image')
plt.show()

# here , the pattern of the image appears random or noise-like.
```

```python
# Part 6 - Think about what each dot product in the matrix multiplication is doing. Describe in your own words what this transformation is doing to the original image.

#In the matrix multiplication of T and img_average_channel, each dot product operation is essentially 
#performing a weighted sum of rows from T based on the values in the corresponding row of img_average_channel. 


```

```python
# Part 7 - We now want to retrieve the original image. Is matrix T invertible? How would you show that it is invertible?

# To check if T in invertible we need to see if its a sqaure matrix i.e it that eqaul number of rows and cols. 

print(T.shape[0]) #rows
print(T.shape[1]) #cols


# we can also see if the rank of matrix is equal to the number of rows to see if its invertible according to linear algebra 

if np.linalg.matrix_rank(T) == T.shape[0]:
        print(np.linalg.matrix_rank(T))
        print("Matrix T is invertible.")
```

```python
# Part 7 - Compute the inverse and store the matrix in Tinv.

Tinv = np.linalg.inv(T) #inverse is calculated with the help of linalg.inv
print(Tinv)
```

```python
# Part 7 - Perform the inverse of the scrambling operation and plot the result to show that you retrieved the original image.

# Compute the inverse operation to retrieve the original image
retrieved_image = np.dot(Tinv, transformed_image)

# Plot the retrieved_image
plt.imshow(retrieved_image) 
plt.title('Retrieved Original Image')
plt.show()
```

```python
# Part 8 - continues - algebra (1)

# Create a numpy array containing numbers from 0 to n-1 using np
col_numbers_array = np.arange(n)

# Shuffle the indices
np.random.shuffle(col_numbers_array)

# Store the shuffled array as "col"
col = col_numbers_array
print(col)

# Create a new matrix CT by rearranging col according to the "col" variable
CT = identity_matrix[col]

# Print the matrix CT
print("Matrix CT:")
print(CT)
```

```python
# Part 8 - continues 


#To perform the same operations as above but on columns we need to perform dot product but in order img * CT and not vice versa to match the size para of img


col_transformed_image = np.dot(img_average_channel,CT)

# To Check the shape of the result
print("Shape of Col_transformed_image:", col_transformed_image.shape)


plt.imshow(col_transformed_image) 
plt.title('Col_Transformed Image')
plt.show()



CTinv = np.linalg.inv(CT)

# Compute the inverse operation to retrieve the original image
col_retrieved_image = np.dot(col_transformed_image, CTinv)


# Plot the retrieved_image
plt.imshow(col_retrieved_image) 
plt.title('Retrieved Original Image based on Column')
plt.show()
```

```python
# Part - Extra Points both row and col scrumbled on avg channel img also done on yellow original img below

# Scramble both rows and columns of the original image using dot product operation twice , ones using the col and ones using the rows
scrambled_image = np.dot(T, np.dot(img_average_channel, CT))


# Plot the retrieved_image
plt.imshow(scrambled_image) 
plt.title('Scrambled Image on both rows and cols')
plt.show()

```

```python
# part - extra points continues 

# To recover the image from scrambled img both on rows and cols we need to again do dot product with inv twice to negate above operations
recovered_image = np.dot(Tinv , np.dot(scrambled_image, CTinv))

# Plot the retrieved_image
plt.imshow(recovered_image) 
plt.title('Recovered Original Image from both rows and col')
plt.show()
```

```python
# Creating arrays of row and column indices for shuffling
cols = np.arange(image.shape[1])

# Shuffling the column indices
np.random.shuffle(cols)

# Initializing the transformed image for yellow/orginal img
ycol_transformed_image = np.zeros_like(image)

# we Looped through each color channel and perform the transformations
for i in range(3):
    # Matrix multiply the color channel with T_cols to scramble columns
    ycol_transformed_image[:, :, i] = np.dot(image[:, :, i], np.identity(image.shape[1])[cols])

# Check the shape of the result by printing it
print("Shape of transformed_image:", ycol_transformed_image.shape)

# lets Plot the transformed RGB image
plt.imshow(ycol_transformed_image.astype(np.uint8))
plt.title("Transformed RGB Image (Columns Shuffled)")
plt.show()
```

```python
# Create an array of column indices that will help us undoing the shuffling
cols_inverse = np.argsort(cols)

# Initializing the recovered image
ycol_recovered_image = np.zeros_like(ycol_transformed_image)

# Loop through each color channel and perform the reverse transformation that will help us retrive the image
for i in range(3):
    # Matrix multiply the color channel with the inverse of T_cols to unshuffle columns
    ycol_recovered_image[:, :, i] = np.dot(ycol_transformed_image[:, :, i], np.eye(ycol_transformed_image.shape[1])[cols_inverse])

# Plot the recovered RGB image
plt.imshow(ycol_recovered_image.astype(np.uint8))
plt.title("Recovered RGB Image")
plt.show()
```

```python
# so now we have created a arrays of row and column indices for shuffling
rows = np.arange(image.shape[0])
cols = np.arange(image.shape[1])

# Shuffling the row and column indices separately
np.random.shuffle(rows)
np.random.shuffle(cols)

# now lets create permutation matrices T for rows and columns separately
T_rows = np.identity(image.shape[0])[rows]
T_cols = np.identity(image.shape[1])[cols]

# then initialize the transformed image
y_col_row_transformed_image = np.zeros_like(image)

# Loop through each color channel and perform the transformations
for i in range(3):
    # Matrix multiply the color channel with T_cols to scramble columns
    y_col_row_transformed_image_cols = np.dot(image[:, :, i], T_cols)

    # Matrix multiply the result with T_rows to scramble rows
    y_col_row_transformed_image[:, :, i] = np.dot(T_rows, y_col_row_transformed_image_cols)

# Check the shape of the result
print("Shape of transformed_image:", y_col_row_transformed_image.shape)

# Plot the transformed RGB image
plt.imshow(y_col_row_transformed_image.astype(np.uint8))
plt.title("Transformed RGB Image")
plt.show()
```

```python
# Now lets do the retrieving of the original image using the same shuffled rows and columns
y_col_row_recovered_image = np.zeros_like(y_col_row_transformed_image)

for i in range(3):
    # Reverse the transformations by using the inverse of T_cols and T_rows
    y_col_row_recovered_image[:, :, i] = np.dot(y_col_row_transformed_image[:, :, i], np.linalg.pinv(T_cols))
    y_col_row_recovered_image[:, :, i] = np.dot(np.linalg.pinv(T_rows), y_col_row_recovered_image[:, :, i])

# Plot the recovered RGB image
plt.imshow(y_col_row_recovered_image.astype(np.uint8))
plt.title("Recovered RGB Image")
plt.show()
```
